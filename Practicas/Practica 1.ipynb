```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def procesar_imagen(image_path, save_prefix):
    img_bgr = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

    cv2.imwrite(f"{save_prefix}_bgr.jpg", img_bgr)
    cv2.imwrite(f"{save_prefix}_gray.jpg", img_gray)


    b, g, r = cv2.split(img_bgr)

    fig, axs = plt.subplots(2, 3, figsize=(15, 10))

    imagenes = [
        (img_rgb, "RGB"),
        (img_gray, "Escala de Grises", "gray"),
        (r, "Canal Rojo", "Reds"),
        (g, "Canal Verde", "Greens"),
        (b, "Canal Azul", "Blues")
    ]

    for i, (img, title, *cmap) in enumerate(imagenes):
        ax = axs[i // 3, i % 3]
        ax.imshow(img, cmap=cmap[0] if cmap else None)
        ax.set_title(title)
        ax.axis("off")

    plt.tight_layout()
    plt.show()

procesar_imagen("ejem1.jpeg", "procesada")
```
