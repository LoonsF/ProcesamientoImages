Parte 3

```python
path = "Gat.jpg"
img = cv2.imread(path, cv2.IMREAD_COLOR)

plt.subplot(141)
plt.title("Imagen Original")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.subplot(142)
plt.imshow(img[:, :, 0], cmap="gray")
plt.title("Canal Azul")

plt.subplot(143)
plt.imshow(img[:, :, 1], cmap="gray")
plt.title("Canal Verde")

plt.subplot(144)
plt.imshow(img[:, :, 2], cmap="gray")
plt.title("Canal Rojo")

print(f"Tamaño de la imagen y canales: {img.shape}")
plt.show()
```

Parte 4

```python
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
r = img[:, :, 2]
g = img[:, :, 1]
b = img[:, :, 0]

rojo_imagen = np.zeros_like(imgRGB)
rojo_imagen[:, :, 0] = r

verde_imagen = np.zeros_like(imgRGB)
verde_imagen[:, :, 1] = g

azul_imagen = np.zeros_like(imgRGB)
azul_imagen[:, :, 2] = b

plt.subplot(141)
plt.title("Imagen Original")
plt.imshow(imgRGB)
plt.axis("off")

plt.subplot(142)
plt.title("Canal Rojo")
plt.imshow(rojo_imagen)
plt.axis("off")

plt.subplot(143)
plt.title("Canal Verde")
plt.imshow(verde_imagen)
plt.axis("off")

plt.subplot(144)
plt.title("Canal Azul")
plt.imshow(azul_imagen)
plt.axis("off")

plt.show()
```

Parte 5

```python
b, g, r = cv2.split(img)

plt.subplot(141)
plt.title("Imagen Original")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(142)
plt.title("Canal Rojo (Falso Color)")
plt.imshow(r, cmap="Reds")
plt.axis("off")

plt.subplot(143)
plt.title("Canal Verde (Falso Color)")
plt.imshow(g, cmap="Greens")
plt.axis("off")

plt.subplot(144)
plt.title("Canal Azul (Falso Color)")
plt.imshow(b, cmap="Blues")
plt.axis("off")

plt.show()
```

Parte 6

```python
path = "Gat.jpg"
img_bgr = cv2.imread(path)
imgRGB = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

print(f"Alto: {img_bgr.shape[0]}")
print(f"Ancho: {img_bgr.shape[1]}")
print(f"Canales: {img_bgr.shape[2]}")

fig, axs = plt.subplots(1, 4, figsize=(20, 5))
axs[0].imshow(imgRGB)
axs[0].set_title("Imagen Original")
axs[0].axis("off")

fig_rojo = axs[1].imshow(img_bgr[..., 0], cmap="gray")
fig.colorbar(fig_rojo, ax=axs[1], shrink=0.5)
axs[1].set_title("Canal Azul (en OpenCV)")
axs[1].axis("off")

fig_verde = axs[2].imshow(img_bgr[..., 1], cmap="gray")
fig.colorbar(fig_verde, ax=axs[2], shrink=0.5)
axs[2].set_title("Canal Verde (en OpenCV)")
axs[2].axis("off")

fig_azul = axs[3].imshow(img_bgr[..., 2], cmap="gray")
fig.colorbar(fig_azul, ax=axs[3], shrink=0.5)
axs[3].set_title("Canal Rojo (en OpenCV)")
axs[3].axis("off")

plt.show()
```

Parte 7

```python
output_path = "Gat.jpg"
cv2.imwrite(output_path + "gato_original.jpeg", img_bgr)
cv2.imwrite(output_path + "gato_rojo.jpeg", rojo_imagen)
cv2.imwrite(output_path + "gato_verde.jpeg", verde_imagen)
cv2.imwrite(output_path + "gato_azul.jpeg", azul_imagen)

print(" Imágenes guardadas correctamente en:", output_path)
```
